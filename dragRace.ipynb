```python
import numpy as np
import random
import matplotlib as mpl
from matplotlib import pyplot as plt
import timeit
from timeit import Timer
%matplotlib inline
```

```python
#Bubble sort, O(N) linear time-space complexity, must iterate through list for each item in it.

#Makes unique list of N items
def makeList(items, lowerBound, upperBound):
    x = []
    
    for i in range(items):
        x.append(random.randint(lowerBound, upperBound))
                 
    return x

#Algorithm
def bubbleSort(listToSort):
    indexLength = len(listToSort) - 1
    case = False
    
    while not case:
        case = True
        
        for i in range(0, indexLength):
            if (listToSort[i] > listToSort[i + 1]):
                case = False
                listToSort[i], listToSort[i + 1] = listToSort[i + 1], listToSort[i]
                
    return listToSort

#Returns data of sorting algo given N number(s) of passes through the algo
def getSortingData(epochs):
    x = 0
    times = []
    
    while x < epochs:
        i = makeList(50, 0, 100)
        
        t = Timer(lambda: bubbleSort(i))
        times.append(t.timeit(number = epochs))
        
        x += 1
        
    return times

#Returns relevant stats of sorting algo time
def getStats():
    
    #Records total sum, average time
    total = 0
    avg = 0
    runs = []
    
    #Call get sorting data to analyze
    x = getSortingData(100)
    
    for i in range(len(x)):
        runs.append(i)
    
    total = sum(x)
    
    avg = total / len(x)
    maxTime = max(x)
    minTime = min(x)
    
    plt.plot(runs, x)
    plt.ylabel('Time (in ms)')
    plt.xlabel('Number of Runs')
    plt.show()
    
    print(f"Average time: {avg}; Maximum Time: {maxTime}; Mininum Time: {minTime}.")
    
    return avg, maxTime, minTime

getStats()

```

```python
#Selection sort
def selSort(listToSort):
    
    for i in range(len(listToSort)):
        minIdx = i
        
        for j in range(i+1, len(listToSort)):
            if (listToSort[minIdx] > listToSort[j]):
                minIdx = j
                
        listToSort[i], listToSort[minIdx] = listToSort[minIdx], listToSort[i]
        
    return listToSort

#Returns data of sorting algo given N number(s) of passes through the algo
def getSortingData(epochs):
    x = 0
    times = []
    
    while x < epochs:
        i = makeList(50, 0, 100)
        
        t = Timer(lambda: selSort(i))
        times.append(t.timeit(number = epochs))
        
        x += 1
        
    return times

#Returns relevant stats of sorting algo time
def getStats():
    
    #Records total sum, average time
    total = 0
    avg = 0
    runs = []
    
    #Call get sorting data to analyze
    x = getSortingData(100)
    
    for i in range(len(x)):
        runs.append(i)
    
    total = sum(x)
    
    avg = total / len(x)
    maxTime = max(x)
    minTime = min(x)
    
    plt.plot(runs, x)
    plt.ylabel('Time (in ms)')
    plt.xlabel('Number of Runs')
    plt.show()
    
    print(f"Average time: {avg}; Maximum Time: {maxTime}; Mininum Time: {minTime}.")
    
    return avg, maxTime, minTime

getStats()
```

```python
#Stack implementation
#Follows last in, first out procedure
#Python's inbuilt list function gets this job done easily

#Declare a list
items = []

#Populate
for i in range(10):
    items.append(i)

#Print to console test
print(items)

#Using inbuild list.pop method, we can see that the last item in is the first item removed
items.pop()

#Print to console test
print(items)

#If we were making our own implementation, it may behave somewhat like this
class stack:
    def __init__(self):
        self.stack = []
        
    def pop(self):
        if len(self.stack) < 1:
            return None
        else:
            return self.stack.pop()
        
    def push(self, item):
        return self.stack.append(item)
    
    def size(self):
        return len(self.stack)
    

        
        

```
